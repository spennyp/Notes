## Types of Bonds
1. Primary - Transfer or sharing of electrons
    * Ionic: 
        * Between metals and non-metals
        * Least electro-negative gives valence electrons to the most electro-negative
        * Non-directional
        * Bond energies between $600-1500 \frac{kJ}{mol}$
    * Covalent: 
        * Between atoms with a small difference in electro-negativity
        * Sharing of valence electrons
        * Directional
        * Wide range of energies and melting points
    * Metallic: 
        * Between metals
        * "Sea of electron" or "cloud"
        * Non-directional
        * Range of bond energies


2. Secondary - No transfer of electrons
    * Van Der Waals:
        * Exists between all atoms or molecules
        * Due to random fluctuations in polarization
        * Weakest type of secondary bonding: $4-30 \frac{kJ}{mol}$
    * Hydrogen:
        * Occurs **between molecules** when H is covalently bonded to **O, N or F** within the molecule
        * Causes difference in electro-negativity of atoms and their spacial arrangement (they become polar)



**Notes**: 

* A directional bond is one where the electrostatic force is unevenly distributed
* Bonding energies are directly proportional to melting temperatures!
* But, dissolution is a different process entirely

# Characterizing Mechanical Properties

![](figs/stressStrain.png)
This shows the **engineering** stress-strain curve from tensile testing. The engineering curve uses the **initial** cross sectional area for stress and strain calculations.

## True vs Engineering Stress and Strain
**Engineering**: based off the initial cross sectional area
   $$\sigma_{eng} = \frac{F}{A_o}; \epsilon_{eng} = \frac{\Delta L}{L_o}$$


**True**: based off the instantaneous cross sectional area
   $$\sigma_{true} = \frac{F}{A}; \epsilon_{true} = \int_{L_o}^{L_i}\frac{1}{L}dL$$
   

**Relating them**: These equations are valid until the onset of necking

\begin{align*}
    \sigma_{true} &= \sigma_{eng}(1+\epsilon_{eng}) \\
    \epsilon_{true} &= \ln{(1+\epsilon_{eng})}
\end{align*}

> Note: The true and engineering stress are the same up until the yield point, they diverge after that, and can be related from yield until ultimate stress with the equation above

## Elastic Deformation

Occurs in the linear region, before the yield stress $\sigma_Y$.
  
  
**Hooke's Law**: $$\sigma = E\epsilon$$  

* $\sigma = \frac{F}{A}$: Stress
* $E$ = slope of linear region: Young's Modulus, which is a measure of material stiffness
    * For materials with similar arrangements of atoms (metals with metals, etc), $\uparrow E = \uparrow T_{melt}$
* $\epsilon = \frac{l_i - l_o}{l_o}$: Strain



**Modulus of Resilience ($MOR$)**: Area under the curve up to $\sigma_Y$  

* This is the recoverable energy absorbed by the material
* $MOR \simeq \frac{\sigma_Y \epsilon_Y}{2} = \frac{\sigma_Y^2}{2E}$: This is if we approximate the area as a triangle
* $[MOR] = Pa = \frac{J}{m^3}$



**Poisson's Ratio**: $$v = -\frac{\epsilon_x}{\epsilon_z} = -\frac{\epsilon_y}{\epsilon_z}$$

* Only valid for *Isotropic* materials (same in all directions), in the elastic regime
* This relates perpendicular contraction to extension caused by tensile stress



**Safety Factor**: We typically design to keep the material from yielding, but add a factor of safety. We want $\sigma_Y \geq N \sigma_w$, where $N$ is the factor of safety, and $\sigma_w$ is the working stress

## Plastic Deformation

After the elastic limit $\sigma_Y$, work hardening occurs until the ultimate stress $\sigma_{UTS}$, at which point necking begins until the fracture stress $\sigma_f$


**Determining $\sigma_Y$**:  
* There are 2 methods, depending the they type of yielding:
![](figs/typesOfYield.png)

    1. Heterogenous yielding: Take $\sigma_Y$ to be the average value after the abrupt fall, at the $\epsilon$ position of the left elbow after the fall 
    2. Homogenous yielding: Use a strain offset method on 0.002, then take a line parallel to the linear region, take $\sigma_Y$ to be where this line intersects the curve



**Elastic Recovery**: Upon unloading plastic deformation, there is elastic recovery from the elastic deformation which has occurred during the elastic region *and some elastic deformation still occurs during plastic deformation*. To find an estimate for the residual strain after deformation we can take the unloading the follow a line parallel to the elastic region.



**Work Hardening:** This is the strengthening of a material which occurs between $\sigma_Y$ and $\sigma_{ULT}$ 

* It is the increase in TRUE stress required to deform a material beyond its elastic limit
* More on this in Module 3 [here](#workHard)


**Necking:** This is the shrinking in cross section which occurs after $\sigma_{ULT}$  

* Caused by cavitation pores (damage) in the material
* In this region there is a complex stress field, it is no longer all axial loading



**Ductility:** This is the amount of deformation a material can tolerate before failure. There are two conventions to measure it (they will have slightly different results)

1. % Elongation $= 100\frac{L_f - L_o}{L_o}$, where $L_o, L_f$ are initial and final lengths respectively 
2. % Area reduction $= 100\frac{A_o - A_f}{A_o}$, where $A_o, A_f$ are initial and final cross sectional areas respectively



**Toughness:** This is the total area under the stress-strain curve up to fracture, and represents the total energy per unit volume absorbed by the material before fracture. 

* Can be approximated (using a rectangle) as: Toughness $= \frac{\sigma_Y + \sigma_{UTS}}{2}\epsilon_f$



**Hardness Testing**: This is a test of the materials resistance to localized plastic deformation. 

* Idea: Place small indent with a known load, and see how far this penetrated into the material
* There are many different scales used, and values are all relative (Rockwell is the most common scale)

> Note: In steel, higher carbon compositions will shift the stress-strain curve up and to the left. 
> * Low carbon steel will have low strength (low $\sigma_{ULT}$), high ductility, and high toughness
> * High carbon steel will have high strength, low ductility, and low toughness



# Crystal Structure and Strengthening Mechanisms

**Some definitions**:

* Crystalline material: atoms arranged in repeating array and exhibit long-range order
* Crystal structure: the spatial arrangement of atoms in crystalline solids (metals, ceramics, polymers)
* Crystal morphology: shape and size of crystals or grains 
    * Crystalline: ordered 
    * Polycrystalline: ordered in islands
    * Amorphous: disordered
* Grain: a bunch of crystals arranged in the same direction
* Grain boundaries: this is where there is an imperfect fit between grains (atomic disorder, and high energy here)

## Unit Cells
A unit cell is a repeating entity that defines the structure of a crystalline solid, there are 3 main types we focus on:

* Body Centered Cubic (BCC)
* Face Centered Cubic (FCC)
* Hexagonal Close Packed (HCP)

![](figs/unitCellTable.png)

**Lattice parameter ($a$)**: Length of the edge of the unit cell

**Coordination number**: number of nearest neighbors or touching atoms in a crystal structure (may need to count the ones from the neighboring cell too)


**Close packed direction**: directions within the unit cell with the highest linear density of atoms
* BCC: Diagonal of cube [111]
* FCC: Diagonal of face [110]


**Atomic packing factor ($APF$)**: volume of atoms in a unit cell per volume of a unit cell 
$$APF = \frac{\text{volume of atoms in unit cell}}{\text{volume of unit cell}} = \frac{\text{# atoms in unit cell} * \frac{4}{3}\pi R^3}{a^3}$$

> Notes: 
> * This is independent of R (atomic radius) for the unit cells we are looking at!
> * May need to do some trig to find a if given R (take the closed pack direction, and count the number of R's on it)
> * For BCC: $a = \frac{4R}{\sqrt{3}}$
> * For FCC: $a = 2\sqrt{2}R$


The figure below gives a better view of what the and HCP lattice looks like
![](figs/HPC.png)


**Theoretical Density**: This is the mass per unit volume of the unit cell if there were not impurities (this is never the case)
$$\rho_{Th} = \frac{n A_i}{V_c N_A}$$
* $\rho_{Th}$ = theoretical density ($\frac{g}{m^3}$)
* $n$ = number of atoms per unit cell
* $A_i$ = atomic weight ($\frac{g}{mol}$)
* $V_c$ = unit cell volume = $a^3$ ($m^3$)
* $N_A$ = Avogadro's number ($6.022 \times 10^{23} \frac{atoms}{mol}$)

## Crystallographic Direction

The crystallographic direction is the way we define a direction within a unit cell. Different directions may exhibit different mechanical properties such as:

* Brittle fracture (occurs on specific planes)
* Elastic properties
* Permanent deformations (occurs by sliding between crystal places)
* Ductility

> These property changes are at the unit cell level, at the macroscopic level we will have the average for a polycrystalline (many crystals, which form many grains)

**Crystallographic Direction Vector**: Defined such that is passes *through the origin* of the coordinate system (CS)

* Specified by three numbers in the notation [$uvw$] (notice no commas between and the square bracket), these should also be reduced to whole numbers!
    * These are the projections on the x, y and z axis of the CS respectively, measured in terms of the unit cell dimensions a, b and c
    * For negative values we use the notation over bar notation, for example is u was a negative number we use [$\bar{u}vw$],
    
![](figs/crystalDir.png)


**Crystallographic Plane**: This specifies a plane within a unit cell, and the notation is ($uvw$). This plane is the plane normal to the vector [$uvw$]
![](figs/crystalPlane.jpeg)


**Planar Density**: This is the number of atoms per unit area of a given crystallographic plane
$$PD = \frac{\text{number of atoms}}{\text{area of plane}}$$

![](figs/planarDensity.png)
Shown above is plane (110) for FCC and BCC, note that both have 2 atoms within the plane (sum the parts), but their areas differ

> Note: 
> * For the number of atoms, this is the equivalent number of **full** atoms in the place (i.e add up the partials)
> * Will likely need to do some trig to get the area of the plane 

## Dislocations

**Defects**: There are 2 types of defects within crystals
![](figs/defects.png)

1. Interstitial: where there is an extra atom
2. Vacancies: where the is a missing atom 

    * All crystals have an equilibrium concentration of vacancies 
    $$ N_v = N \exp(\frac{-Q_v}{RT}) $$
    > Just below melting point, $\frac{N_v}{N} \approx \frac{1}{10000}$
    
        * $N_v$: Number of vacancies per unit volume
        * $N$: Number of atoms per unit volume (as if there were no vacancies
        * $Q_v$: activation energy for vacancy formation
        * $R$: Boltzmann's constant (1.38 $\frac{J}{atom*K}$)


 
**Diffusion**: Atoms move between vacancies, and interstitial atoms move around (happens all the time)
* Allows atoms to rearrange to obtain lower energy configurations
* Important in creep, recrystallization, grain growth and phase transformations



**Dislocations**: Linear crystalline defects whos motions give rise to *plastic deformation* in crystalline materials
* These disrupt the lattice structure, causing causing *elastic stress and strain fields*
![](figs/dislocationDisrupt.png)
* Dislocation can interact with each other, and cancel each other out
![](figs/dislocationCancel.png)


**Dislocation Density**: Measures the length of dislocations in a unit volume of the crystal
$$\rho = \frac{\text{Total length of dislocation line}}{\text{Volume}}; [\rho] = \frac{m}{m^3}$$


**Slip**: Process by which plastic deformation is produces by dislocation movement from applied *shear stress* which causes *severing and reformation* of atomic bonds
* Slip plane: the plane along which this the dislocation movement occurs (this is a crystallographic plane)
![](figs/slip.png)


**Slip System**: Describes the preferred slip plane, and direction (slip direction) for dislocation movement
* Preferred slip plane: the plane with **densest atomic packing**
* Preferred slip direction: the direction with the **highest linear density**
* Several slip systems can exist


**Slip in a single crystal material**: When a polished metal surface is plastically deformed, slip lines appear on the surface, which indicates shearing across atoms.
* Even if the applied stress is purely tensile, shear components exist, which are called resolved shear stress
$$\tau_R = \sigma \cos(\phi) \cos(\lambda)$$
    * $\tau_R$: resolves shear stress
    * $\sigma$: applied stress
    * $\phi$: angle between normal to slip plane and applied stress direction
    * $\lambda$: angles between the slip and stress directions
* Slip occurs when: $\tau_{R,max} = \sigma(\cos(\phi) \cos(\lambda))_{max} \geq \tau_{crss}$, where $\tau_{crss}$ is the critical shear stress
* Thus, the yield stress is: $$\sigma_Y = \frac{\tau_{crss}}{(\cos(\phi) \cos(\lambda))_{max}}$$
![](figs/resolvedShear.png)

> Note: While slip involves severing atomic bonds, we notice that the required yield stress is lower than the force to break all the atomic bonds, this is because not all bonds on the slip planes are severed simultaneously, instead they are broken in one place at a time


**Slip in a polycrystalline material**: These are stronger than their single crystal counterparts. The random crystallographic orientation of grains means the direction of slip varies grain to grain (has to zig-zag)


## Strengthening of Crystals

Because dislocation motion is the reason for plastic deformation, materials can be strengthened by impending dislocation motion.

There are 3 main ways to strengthen a material:
1. **Grain size reduction**: Grain boundaries impede dislocation motion as the dislocations will need to change directions (discontinuity of slip planes)
    * Hall-Petch equation: Relates yield strength and grain size, reducing grain size increases yield strength
        $$\sigma_Y = \sigma_o + k_yd^{-0.5}$$
       * $d$: the average grain diameter
       * $\sigma_o$ and $k$: material constants 

    

2. **Solid-solution**: Have a material with at least 2 different types of atoms (i.e create an alloy)
    * Substitute solute: alloy atom has similar size to host atom and replaces some of the host in the crystal structure
        * Smaller substitute atoms create a tensile stress, where larger create compressive. These offset the stress from dislocations
    * Interstitial solute: alloy atom is smaller than host atom and fits in between some of the host atoms in the crystal structure
![](figs/solidSolution.png)



<a id='workHard'></a>
3. **Work Hardening**: As we increase past yield stress, work hardening occurs, which strengthens the material. 
* This occurs due to the increase in dislocation, and dislocation-dislocation strain interactions make dislocation motion more difficult
* The TRUE stress up to necking can be approximated by (after yielding): 
    $$\sigma_{true} = K (\epsilon_{true})^n$$
    * $K$: strength coefficient
    * $n$: work hardening exponent
* Expressed as percent cold work:
    \begin{align*}
         \% CW &= 100\frac{A_o-A_d}{A_o} \\
         &= 100\frac{t_o - t}{t_o}
    \end{align*}
    * Called cold working, because this only occurs for pure metals at $T < 0.4 T_m$, and for alloys $T < 0.6 T_m$, where $T_m$ is the melting temperature. ALL IN **K**
* One method to work harden is rolling
* There is a trade off here, work hardening increases strength, but decreases ductility
![](figs/effectOfWH.png)

## Recovery, Recrystallization and Grain Growth

Annealing heat treatment can reverse the effects of work hardening (cold working). The stored energy from work hardening is the driving force in this 3 step process (recovery, recrystallization, grain growth)

![](figs/annealing.png)


1. **Recovery**: release of some of the stored energy by dislocation motion (without applied stress) by atomic diffusion at high temperatures
    * Restoration of electrical and thermal properties to pre-cold work
    * Tensile stress slightly decreases, and ductility slightly increases

    
2. **Recrystallization**: formation of new, strain free, and equiaxed (equal dimensions in all directions) grains
    * Driving force is stored internal energy 
    * Throughout this process, there is a significant decrease in strength, and increase in ductility
    * Recrystallization temperature: the temperature required which recrystallization is complete in 1 hour
        * This decreases with increased cold work (i.e **rate of recrystallization increases as cold work increases**)
            ![](figs/recrystalizationRate.png)
    * Thermally activated process:
        $$\text{rate} = \frac{1}{t_{recrx}} = A \exp(\frac{-Q}{RT})$$
        * $t_{recrx}$: recrystallization time (i.e time for recrystallization to be 100% complete)
        * $Q$: activation energy (decreasing function of amount of cold work)
        * $T$: temperature (K)
        * $A$: constant
    * Occurs more rapidly in pure metals than alloys (due to alloys reducing motion of grain boundaries)
    
3. **Grain growth**: reduces grain boundaries (and thus energy, as boundaries are regions of high energy) within the material by larger grains consuming smaller ones
    * For many polycrystalline materials, the equation that describes the grain size with time is:
        $$d^n - d_o^n = Kt$$
        * $d_o$: initial grain size
        * $d$: grain size after time t
        * $K, n$: constants




> Note: If there is not a sufficient amount of cold work, recovery and recrystallization will not being (but grain growth can still occur

## Crack Propagation

* Material defects contribute to fracture
* At cracks, there are **stress concentrations** which can lead to material failure

    ![](figs/crack.png)
    * $a$: half the length of an internal crack, or full length of a surface crack

* **Maximum crack tip stress (stress concentration factor $K_t$)**:
\begin{align*}
    \sigma_m &= 2 \sigma_0 \sqrt{\frac{a}{\rho_t}} = K_t \sigma_0 \\
    K_t &= 2 \sqrt{\frac{a}{\rho_t}} = \frac{\sigma_m}{\sigma_0}
\end{align*}
    * $\sigma_m$: maximum stress concentration
    * $\sigma_0$: nominal stress (as if crack wasn't there)
    * $\rho_t$: radius of curvature of crack tip
    * $K_t$: stress concentration factor, this is a measure of stress amplification at the crack tip
    
    
* **Griffith Brittle Fracture**: crack will propagate when release of strain energy exceeds energy required to form a new crack
$$\sigma_c = \sqrt{\frac{2E \gamma_s}{\pi a}}$$
    * $\sigma_c$: critical stress for crack propagation in brittle material
    * $E$: elastic (Young's) modulus
    * $\gamma_s$: specific surface energy
    
    
* **Fracture Toughness ($K_{Ic}, K_I$)**: material property that is a measure of the resistance to brittle fracture when a crack is present
    * There are 3 modes of loading, each with their own $K_c$ ($K_{Ic}, K_{IIc}, K_{IIIc}$, we only really discuss mode 1):
        ![](figs/kcmodes.png)
    \begin{align*}
        K_{Ic} &= Y \sigma_c \sqrt{\pi a_c} \implies \sigma_c = \frac{K_c}{Y \sqrt{\pi a_c}} \\
        K_I  &= Y \sigma \sqrt{\pi a} \\ 
        K_I &> K_{Ic} \implies \text{crack will propagate}
    \end{align*}
    * $K_{Ic}$: fracture toughness (mode I)
        * Fracture toughness decreases with increasing yield strength 
    * $K_I$: stress intensity factor (sometimes just denoted as $K$)
    * $Y$: crack geometry and load dependent constant
        * Typically determined from a plot of $\frac{a}{W}$ vs $Y$
    * $\sigma_c$: critical stress for failure
    * $a_c$: **critical crack length**, this is the length of the crack that will cause failure for a given Y
    
  
* **Comments**:
    * Can drill holes at crack tips to increase the radius of curvature, thus decreasing stress concentration
    * Can detect cracks with ultrasonic methods

## Fatigue

* Failure occurring below yield stress from long times of cyclic **tensile** loading, which causes crack nucleation and growth
    * Failure occurs when crack grows to critical length $a_c$, i.e $K_c$ is exceeded

* Single largest cause of failure!

* Benchmark: lines in the material (like a tree trunk) which indicates where crack propagation occurred (trace them back), spacing can indicate rate

* **The 3 stages**:
    1. Initiation
    2. Propagation
    3. Final fracture ($K=K_c$), the critical crack length has occurred

* **Cyclic Loading**:
![](figs/cyclicLoading.png)
\begin{align*}
    \text{mean stress} &= \sigma_m = \frac{\sigma_{max} + \sigma_{min}}{2}  \\
    \text{stress range} &= \sigma_r = \sigma_{max} - \sigma_{min} \\
    \text{stress amplitude} &= \sigma_a = \frac{\sigma_{max} - \sigma_{min}}{2} \\
    \text{stress ratio} &= R = \frac{\sigma_{max}}{\sigma_{min}}
\end{align*}


* **S-N curve**: stress amplitude $S$ vs number of cycles $N$ curve, which indicates fatigue life
    * Fatigue limit: the stress amplitude at which fatigue failure does not occur (only present in some materials)
    * Sometimes multiple curves are give, where each is with a different probability of failure
    ![](figs/SN.png)
    
* **Extending fatigue life**:
    1. Surface treatment:
        1. Shot Peening: hitting surface with high speed particles, leaving residual compressive stressed which delay crack initiation
        2. Case hardening: hardens the surface, while allowing the metal underneath to remain tough (high $K_c$)
    2. Mechanical design: reduce stress concentrations (ex. removing sharp corners, blunt crack tips)

## Ductile to Brittle Transition (DBT)

* At low temperatures, there can be a dramatic reduction in material ductility
    * DBT does not typically occur in FCC and HCP metals
* **DBT Temperature (DBTT)**: This is the temperature at which ductile to brittle transition happend
    * Yield strength is unchanged, but elongation and thus toughness (area under stress-strain curve) is
    * At lower temperatures, movement of dislocations (ductile failure mode) in BCC metals are more difficult, and the fracture mode switching from ductile to brittle as the propagation of cracks becomes easier than dislocations
* **Strain rate**: this is the rate of material deformation, higher rates will shift the DBT curve to the right (i.e increasing DBTT, thus is brittle for larger range of T)  
    ![](figs/DBT.png)
    ![](figs/fractureSurface.png)

* **Charpy Impact Test**: test to simulate the severe service conditions where the DBT is important so we can study brittle material fracture (can't be determined from stress-strain curve, it is from propagation of defects). Measures the absorbed energy of the specimen
    * Not a material property, it is qualitative and not directly transferable to design
    * Conditions that are simulated
        1. High strain rate
        2. Lower temperature
        3. Triaxial stress state (non uniform)
    * Test: 
        1. Drop hammer from know height
        2. Hammer strikes the specimen (notch to create triaxial stresses, and protect hammer)
        3. Final hammer position is measured
        4. Energy absorbed by specimen can be determined with conservation of energy (neglecting friction)
            * Brittle material: low Charpy impact energy
            * Fracture surface: clean and shiny is brittle, dull is ductile 
    ![](figs/charpy.png)
    
* **Effects of carbon content on DBTT**: more carbon means a higher DBTT, higher yield strength, but lower DBT fracture energy
![](figs/DBTTCarbon.png)



## Creep

* **Time dependent** and permanent deformation of a material subject to **constant load at high temperatures**
    * Above $0.4T_m$ for metals
    * Higher temperatures and higher load mean fasture fracture time (higher strain rate)
    * Occurs from sliding grains (not disloactions like at low $T$)

* **3 stages of creep**: 
<img src="figs/creepCurve.png" style="width: 500px;"/> 

    1. Primary: decreasing creep rate due to work hardening
    2. Seconary: constant creep rate $\epsilon_s$ due to balancing of work hardening and recovery
    3. Tertiary: acceleration of creep rait until rupture

* **Components of strain**: elastic, plastic, creep strain respectivly. 
    $$\epsilon_{total} = \epsilon_E + \epsilon_P + \epsilon_C$$
    * Elastic and plastic depending on loading, where as creep strain is time dependent
    <img src="figs/totalStrain.png" style="width: 500px;"/>

* **Strain rate in secondary region**:
    $$\dot{\epsilon_s} = K_2 \sigma^n \exp \left(\frac{-Q_c}{RT} \right)$$
    * $\dot{\epsilon_s}$: creep rate in secondary stage ($\frac{d \epsilon_s}{dt}$)
    * $K_2$: material constant
    * $n$: creep exponent (material constant)
    * $Q_c$: activation energy for creep
    * $R$: gas constant ($2.31$ J mol/K)
    * $\sigma$: tensile or compressive stress

* **Creep extrapolation (Larson Miller)**: extrapolate data from higher temperatures and shorter times to apply to practical timescales of years
    $$m = T(C+\log(t_r))$$
    * $m$: Larson miller parameter, this stays constant
    * $C$: constant
    * $T$: operating temperauture (K)
    * $t_r$: time to rupture in **hours**
    
* **Factors affecting creep characteristics**:
    * **Higher melting temperature**, **elastic modulus** and **larger grain size** are more resistant to creep
        * Large grains result in smaller surface area for sliding grain boundaries which means less creep
        * **Single crystals** are desirable for this reason
        

## Corrosion 

* **Process**: Electrochemical attack resulting in oxidation that converts refined metal to a more chemically stable form (lower energy)
    * Oxidation: loss of electrons during a reaction
    * Elecrochemical cell forms where one metal surface serves as the anode (+), and another as the cathode (-)
        * Anode gives electrons to the cathode
    * Rust flakes off as the ocide product has a greater volume and poor adhesion to the metal  

    <img src="figs/corrosion.png" style="width: 500px;"/>

* **4 ingredients of corrosion**:
    1. Reduction in energy: thermodynamic driving force
    2. Water or moisture to act as a medium for ionic transport
    3. Sorce of oxidant (normally $O_2$)
    4. Mass transport of oxidant to the metal surface
    
* **Types of corrosion**:
<img src="figs/typesOfCorrosion.png" style="width: 500px;"/>  

    * Stress corrosion: corrosion can promote crack propogation in components under stress

* **Galvanic corrosion (not shown above)**: when 2 different metals or alloys are **electrically coupled**, and **exposed to an electrolyte**
    * More reactive metal acts as anode and is oxidized (loses electrons and corrodes). Less reactice metal acts as the cathode, and is protected from corrosion
    * **Galvanic series** is very important, and we can use this to help protect material from corrosion by having a sacraficial reactive (anodic) material:  

    <img src="figs/galvanicSeries.png" style="width: 500px;"/>
    
    
* **Corrosion protection for metals**: can discourage by removing reactants, reducing temperature, using inhibitors in fluids, and altering the design. But there are 3 main ways used for deliberate protection:
    1. Self driven formation of a protective coating: when oxide layer protects against further corrosion (stainless steel)
    2. Physical barriers (paint, noble coating): uses a less chemically active layer, but prone to physical damage rendering the film useless
    3. Sacrificial or cathodic protection: uses more chemically active metal as a sacrifice anode, this sacrificial material needs to be replaced


# Microstructure

* Refers to the material stucture at the microscopic level
* Different microstructures can have different material properties which we can take advantage of

## Phase Diagrams

* Allows us to see how microstructure changes with different material compositions under different temperature and pressure conditions
    * Used to study: alloying, heat treatments, microstructure
    * Depict **EQUILIBRIUM CONDITIONS ONLY**, changes are assumed to be quisistatic (very slow)

* **Phase**: homogeneous state of a system that has uniform physical and chemical properties (ex: solid, liquid, gas, allotropes)
    * **Allotrope**: solids with the same composition, but different microstructure (FCC, BCC, HPC), these are considered different phases. We denote the solid allotropes with greek symbols ($\alpha, \beta, \gamma$)
    * Liquid: $L$

* **Types of phase diagrams**:
    1. Uniary: single composition (P vs. T)
    2. Binary: wt% composition of materials (T vs wt%)
        * Isomorphous: liquid and solid region each contain only one phase
        <img src="figs/isomorphous.png" style="width: 500px;"/>

        * Eutectic: multiple different solid phases  
        <img src="figs/eutectoid.png" style="width: 500px;"/>


* **Anatomy of phase diagram**: 
    * Liquidus line: boundary between 100% liquid, and a liquid/solid two phase region
    * Solidus line: boundary between 100% solid, and a liquid/solid two phase region
    * Solvus line: boundary between 100% $\alpha$ and a $\alpha, \beta$ two phase (two allotrope) region 
    * Eutectic point ($E$): point where two liquidus lines meet (can go $L \to \alpha, \beta$)
    * Eutectoid point: point where two solvus lines meet (can go $\gamma \to \alpha, \beta$)
    * Eutectic/eutectoid temperature/isotherm: temperature at which the eutectic/eutectoid point occurs

* **Phase composition**:
    * Single phase region: composition is as indicated on the $X$ axis ($C_0$)
    * Two phase region: use a **tie line**, the intersection of this line with the phase boundaries gives the respective composition
        * Analogy is a slushy melting: 50% solid, 50% liquid, but liquid has higher % composition of sugar

        <img src="figs/phaseComposition.png" style="width: 500px;"/>

* **Phase fraction**: what mass fraction of each phase is present
    * Single phase region: 100% that phase
    * Two phase region: use a **tie line** and use the **lever rule** (opposite arm over total line length)
        * Example: for the image above the mass fraction liquid ($W_L$) and alpha solid ($W_\alpha$) are 
        \begin{align*}
             W_L &= \frac{S}{R+S} = \frac{C_\alpha - C_0}{C_\alpha - C_L} \\
             W_\alpha &= \frac{R}{R+S} = \frac{C_0 - C_L}{C_\alpha - C_L} \\
        \end{align*}
        * **Mass conservation**: $W_L + W_\alpha = 1$

* **Microstructural changes from cooling**: 
    * Equilibrium cooling moves vertically down the phase diagram, compositions can be calculated with the lever rule, non-equilibrium gives rise to non-uniform cored structures and we can't use the lever rule
    * There are multiple possible regions that can get transversed during cooling, examples are at $C_1, C_2, C_3, C_4$  

     <img src="figs/phaseCoolingRegions.png" width="350px" />

    1. **Doesn't cross solvus line**: example $C_1$  
        <img src="figs/C1.png" width="350px"/>  

        * Only $\alpha$ solids and liquid exists
        * Results in an alloy of **uniform** $C_1$ composition as its cooled towards $0K$
        
    2. **Crosses solvus line, but not eutectic isotherm**: example $C_2$ 
        <img src="figs/C2.png" width="350px" /> 

        * Uniform $\alpha$ after crossing the through $\alpha$, liquid region
        * Upon crossing the solvus line, $\beta$ particles form, and grow slightly as $T$ lowers further
        * Non-uniform 
        
    3. **Crosses eutectic point**: example $C_3$  
        <img src="figs/C3.png" width="350px" /> 

        * Upon crossing the eutectic isotherm, $L \to \alpha, \beta$ instantly (no in between)
        * **Lamellae** form due to atomic diffusion, which is fine altering layers of $\alpha, \beta$ solids
        * Upon further cooling, there is minimal change, but $\beta$ increases slightly
        
    4. **Crosses both solidus line, and eutectic isotherm**: example $C_4$
        <img src="figs/C4.png" width="350px" />

        * This is called **hypo** or **hyper** eutectic if the composition is before or after the eutectic composition respectively 
        * Upon crossing the solidus line, $\alpha$ or $\beta$ blobs form with the liquid, these are called primary $\alpha$ or $\beta$ (former is hypo, latter is hyper)
        * Upon crossing the eutectic isotherm, lamellae layers form around the blobs from the liquid portion of the composition, while the solid portion right before and right before the isotherm remains. 
            * The solid that is the same phase as the primary is called eutectic ($\alpha$ is hypo, $\beta$ is hyper)
        * Can draw a tie-line to find composition right before, and right after the isotherm


## Steel Microstructures

* Mechanical properties are dependent on carbon content, and microstructural transformation upon cooling
* **Phase boundaries of microstructure**: phase boundaries resist dislocation movement
    * More phase boundaries $\implies$ stronger
    * Less phase boundaries $\implies$ more ductile
* $Fe-Fe_3 C$ phase diagram (relevant part and **equilibrium cooling only**):  

    <img src="figs/fePhase.png" style="width: 600px;"/>  

    * **Allotropes**:
        * $\alpha$: **Ferrite** (BCC), this is soft and ductile
        * $\gamma$: **Austenite** (FCC)
        * $Fe_3C$: **Cementite** (vertical line at 6,7 wt% C), this is brittle and hard
    * **Pearlite**: formed from equilibrium cooling at eutectoid composition (0.76 wt% C)
        * $\gamma \to \alpha + Fe_3C$ (pearlite), ferrit + cementite from crossing eutectoid point
            * Lamellar mircrostructure from diffusion over instantaneous transition 
            * In between the soft, ductile ferrite and brittle, hard cementite
        * Equilibrium hypo or hyper eutectoid cooling: $\gamma \to \alpha + Fe_3C$, with pro-eutectoid (primary) parts which form at grain boundaries of $\gamma$ as they form
            * Hypo will have pro-eutectoid (primary) ferrite (portion from before isotherm), and pearlite
                <img src="figs/hypoSteel.png" style="width: 300px;"/>  
                <img src="figs/hypoSteelScan.png" style="width: 400px;"/>  

            * Hyper will have pro-eutectoid (primary) cementite and pearlite
                <img src="figs/hyperSteel.png" style="width: 300px;"/>  
                <img src="figs/hyperSteelScan.png" style="width: 400px;"/> 
        * **The faster the cooling rate, the thinner the inter-laminar spacing (i.e more phase boundaries)**

* **Heat Treatment**: if we heat to single phase austenite ($\gamma$), we can vary cooling rate to manipulate microstructure
    1. Autenitize: heat to temperature region where 100% austenite $\gamma$ is formed
    2. Quench: use water or oil to rapidly cool to form non-equilibrium, hard microstructures
    3. Temper: reheat to $200-550 \deg C$ to decrease hardness and increase ductility (bring some life back to the material)
